```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
import seaborn as sns


# In[2]:


#Importing BREAST CANCER DATA FROM SKLEARN DATASETS
data = load_breast_cancer() 
print(data.keys())


# In[3]:


print(data.feature_names,"\n")  #column names


# In[4]:


data.target[0:20]  #target array
print(data.target[0:20])

# In[5]:


data.target_names


# In[6]:


df=pd.DataFrame(data.data,columns=data.feature_names)
df.size


# In[8]:


df.shape


# In[10]:


df.head(3)


# In[11]:


df['Target']=data.target
print("\n","\n","\n",df['Target'])


# In[12]:


df.head()


# In[13]:


df.info()


# In[18]:


df.dtypes


# In[19]:


#WE WILL USE LOGISTIC REGRESSION FOR CLASSIFICATION¶

#IMPORTING LIBRARIES FROM SKLEARN
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics


# In[20]:


X=data.data
y=data.target
X_train,X_test,y_train,y_test=train_test_split(X,y)


# In[21]:


LR=LogisticRegression()


# In[22]:


LR.fit(X_train,y_train)


# In[24]:


#MODEL ACCURACY
LR.score(X_test,y_test)  # Generates 93% accuracy using LOGISTIC REGRESSION MODEL
#Generated accuracy of 93%


# In[25]:


yhat=LR.predict(X_test)   #Predicts labels for test data
yhat


# In[27]:


#LOGISTIC REGRESSION MODEL ACCURACY METRIC: CONFUSION MATRIX
from sklearn.metrics import classification_report, confusion_matrix
CM=metrics.confusion_matrix(y_test,yhat,labels=[0,1])


# In[28]:


print(CM)


# In[29]:


plt.figure(figsize=(6,6))
sns.heatmap(CM,annot=True)


# In[31]:


classification_report(y_test,yhat)
#Shows PRECISION, RECALL, F1 SCORE


# In[ ]:




```
