```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


from sklearn.datasets import load_iris


# In[2]:


iris = load_iris()


# In[3]:


print(iris.data[:5])
X = iris.data
print("\n",type(X))


# In[4]:

#print()
print("\n",iris.feature_names)


# In[5]:


#Printing of target label values or prediction label values
print("\n","Printing of target label values or prediction label values")
print(iris.target,"\n")
y=iris.target


# In[7]:

#Printing of target label names or prediction label names
print("#Printing of target label or prediction label")
print(iris.target_names,"\n")


# In[8]:

#check number of rows and columns in dataset
print(iris.data.shape,"\n")


# In[9]:


#splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 4)


# In[10]:


print(X_train.shape)

print(X_train,"\n")

print(y_train.shape)

print(y_train,"\n")

print(X_test.shape)

print(X_test,"\n")

print(y_test.shape)

print(y_test)



# In[11]:


from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
k_range = range(1,26)
scores = {}
scores_list = []

for k in k_range:
    # Create KNN classifier and set ‘n_neighbors’ to some value
    knn = KNeighborsClassifier(n_neighbors = k)

    # Fit the classifier to the data
    knn.fit(X_train,y_train)

    #This returns an array of predicted class labels for each sample in X_test, which is assigned to the variable y_pred.
    y_pred = knn.predict(X_test)



    #This code calculates the accuracy of a machine learning model by comparing the predicted values (stored in y_pred)
    #with the actual values (stored in y_test).

    scores[k]= metrics.accuracy_score(y_test,y_pred)

    #print(scores[k])
    scores_list.append(scores[k])


# In[12]:

print()
print("\n",scores_list)


# In[13]:


get_ipython().run_line_magic('matplotlib', 'inline')
import matplotlib.pyplot as plt
plt.plot(k_range,scores_list)
plt.xlabel("k value")
plt.ylabel("Testing Accuracy")


# In[ ]:




```
